```python
import pandas as pd
```

```python
# The path to our CSV file
gunviolence_file = "/Users/mariaquinton/Desktop/Project-2---Group-E/datasets/reduced_gun_violence.csv"
firearms_file = "/Users/mariaquinton/Desktop/Project-2---Group-E/datasets/reduced_firearms.csv"
```

```python
# Read our data into pandas
violence_df = pd.read_csv(gunviolence_file)
violence_df.head()
```

```python
violence_df['date'] = pd.to_datetime(violence_df.date)
violence_df
```

```python
firearms_df = pd.read_csv(firearms_file)
firearms_df.head()
```

```python
firearms_df['month'] = pd.to_datetime(firearms_df.month)
firearms_df.head()
```

```python
firearms_df = firearms_df.rename(columns={'month': 'date'})
```

```python
firearms_df.head()
```

```python
# Merge the two dataframes on Date
merged_df = pd.merge(violence_df, firearms_df, on="date", suffixes=("_violence", "_firearms"))
merged_df.head()
```

```python
merged_df.to_html('/Users/mariaquinton/Desktop/Project-2---Group-E/datasets/combined_data.html')
```

```python
merged_df.to_csv('/Users/mariaquinton/Desktop/combined_data.csv')
```
